```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

try:
    image = cv2.imread('noisy.png', cv2.IMREAD_GRAYSCALE)
    if image is None:
        raise FileNotFoundError("dolphin.png 파일을 찾을 수 없습니다.")
except Exception as e:
    print(e)
    exit()
```

```python
# 이미지가 성공적으로 로드된 경우에만 필터링을 진행합니다.
if image is not None:
    # --- 1. Average filter 적용 ---
    kernel_sizes = [3, 5, 7, 9]
    avg_filtered_images = []
    for k in kernel_sizes:
        # cv2.blur() 함수를 이용해 Average filter 적용
        filtered = cv2.blur(image, (k, k))
        avg_filtered_images.append(filtered)

    # --- 2. Gaussian filter 적용 ---
    sigma_values = [0.5, 1, 2, 4]
    gaussian_filtered_images = []
    for s in sigma_values:
        # cv2.GaussianBlur() 함수를 이용해 Gaussian filter 적용
        filtered = cv2.GaussianBlur(image, (9, 9), s)
        gaussian_filtered_images.append(filtered)

    # --- 3. 결과 시각화 ---
    plt.figure(figsize=(24, 12))

    for i, img in enumerate(avg_filtered_images):
        # subplot(전체 행, 전체 열, 현재 위치)
        plt.subplot(2, 4, i + 1)
        plt.imshow(img, cmap='gray')
        plt.title(f'Average Filter ({kernel_sizes[i]}x{kernel_sizes[i]})', fontsize=16)
        plt.axis('off')

    for i, img in enumerate(gaussian_filtered_images):
        plt.subplot(2, 4, i + 5)
        plt.imshow(img, cmap='gray')
        plt.title(f'Gaussian Filter (9x9, sigma={sigma_values[i]})', fontsize=16)
        plt.axis('off')

    plt.tight_layout(pad=2.0)
    plt.show()
```
